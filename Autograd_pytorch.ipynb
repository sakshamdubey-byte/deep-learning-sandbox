AAM ZINDAGI - doing manuallly


```python
import torch

# 1. Initialize input (x), weight (w), bias (b), and target (y)
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad=False) # No autograd tracking needed
b = torch.tensor(1.0, requires_grad=False)

# 2. Forward Pass: Compute prediction and loss
# Equation: y_pred = w * x + b
y_pred = w * x + b

# Loss function (Squared Error): L = (y_pred - y)^2
loss = (y_pred - y) ** 2
print(f"Manual - Initial Loss: {loss.item()}")

# 3. Backward Pass: Manual Gradient Computation (Calculus)
# dLoss/dy_pred = 2 * (y_pred - y)
# dy_pred/dw = x
# By Chain Rule: dLoss/dw = (dLoss/dy_pred) * (dy_pred/dw)
dloss_dw = 2 * (y_pred - y) * x

print(f"Manual Gradient (dloss/dw): {dloss_dw.item()}")

# 4. Update weight using Gradient Descent (Learning Rate = 0.01)
lr = 0.01
w = w - lr * dloss_dw
print(f"Manual - Updated Weight: {w.item()}\n")
```

MENTOS ZINDAGI


```python
import torch

# 1. Initialize variables (Turn on requires_grad=True for the weights we want to optimize)
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad=True) # PyTorch will now track operations on 'w'
b = torch.tensor(1.0, requires_grad=True)

# 2. Forward Pass: Same mathematical operations
y_pred = w * x + b
loss = (y_pred - y) ** 2
print(f"Autograd - Initial Loss: {loss.item()}")

# 3. Backward Pass: Let PyTorch do the heavy lifting
# This computes gradients for any tensor that has requires_grad=True
loss.backward()

# Look at the accumulated gradient in the .grad attribute of the weight
print(f"Autograd Gradient (w.grad): {w.grad.item()}")

# 4. Update weight
# We wrap this in torch.no_grad() so PyTorch doesn't track the update step itself
lr = 0.01
with torch.no_grad():
    w -= lr * w.grad

print(f"Autograd - Updated Weight: {w.item()}")
```

CLEARING OUT GRADIENT AFTER USE

```python
import torch

x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad=True)

# Iteration 1
loss1 = (w * x - y) ** 2
loss1.backward()
print(f"Iteration 1 - Gradient: {w.grad.item()}") # Prints -2

# CRITICAL STEP: Clear the gradient manually before the next pass
w.grad.zero_()

# Iteration 2
loss2 = (w * x - y) ** 2
loss2.backward()
print(f"Iteration 2 - Fresh Gradient: {w.grad.item()}") # Correctly prints -2 because the cache was cleared!
```

```python
import torch

# Create a tensor with gradient tracking enabled
x = torch.tensor(2.0, requires_grad=True)
print(f"Initial x: {x}")

# --- Option 1: In-place modification using requires_grad_() ---
# This permanently changes the tensor's setting
x.requires_grad_(False)
print(f"Option 1 - x after requires_grad_(False): {x}")

# Resetting x for the next options
x.requires_grad_(True)

# --- Option 2: Using detach() ---
# This creates a new view of the tensor that shares the same data but does not require gradients
y = x.detach()
print(f"Option 2 - original x: {x}")
print(f"Option 2 - detached y: {y}")

# --- Option 3: Using the torch.no_grad() context manager ---
# This is the most common approach used during model evaluation or weight updates
with torch.no_grad():
    z = x ** 2
    print(f"Option 3 - Inside no_grad block, z requires grad? {z.requires_grad}")

# Outside the block, tracking resumes normally
w = x ** 2
print(f"Outside no_grad block, w requires grad? {w.requires_grad}")
```
