```python
import torch
import time

# 1. Define the size of the square matrices
# A size of 5000x5000 creates a heavy mathematical workload to really test the hardware.
matrix_size = 20000

print(f"Generating random {matrix_size}x{matrix_size} matrices...")

# 2. Benchmark on CPU
# We create random matrices on the host memory (RAM) and compute on the CPU
matrix_cpu1 = torch.randn(matrix_size, matrix_size)
matrix_cpu2 = torch.randn(matrix_size, matrix_size)

print("Starting CPU calculation...")
start_time_cpu = time.time()

# Perform matrix multiplication on CPU
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)

cpu_duration = time.time() - start_time_cpu
print(f"Time taken on CPU: {cpu_duration:.4f} seconds\n")


# 3. Benchmark on GPU
# We check if a CUDA-capable GPU (like a Tesla T4) is available
if torch.cuda.is_available():
    print(f"GPU Found: {torch.cuda.get_device_name(0)}")

    # Move the matrices from system RAM over to the GPU VRAM
    matrix_gpu1 = matrix_cpu1.to('cuda')
    matrix_gpu2 = matrix_cpu2.to('cuda')

    print("Starting GPU calculation...")
    start_time_gpu = time.time()

    # Perform matrix multiplication on GPU
    result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)

    # CRITICAL STEP: PyTorch executes GPU operations asynchronously.
    # torch.cuda.synchronize() forces Python to wait until the GPU completely finishes
    # the math before recording the stop time.
    torch.cuda.synchronize()

    gpu_duration = time.time() - start_time_gpu
    print(f"Time taken on GPU: {gpu_duration:.4f} seconds\n")

    # 4. Compare Results
    speedup = cpu_duration / gpu_duration
    print("-" * 40)
    print(f"⚡ RESULT: The GPU is {speedup:.2f}x FASTER than the CPU! ⚡")
    print("-" * 40)
else:
    print("❌ No GPU detected! Please go to Runtime -> Change runtime type and select a GPU.")
```
