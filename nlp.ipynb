```python
# REMOVING STOP WORDS
import nltk
from nltk.corpus import stopwords

# Download the missing dataset
nltk.download('stopwords')

# Now this will work perfectly!


def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return " ".join(x)

print(remove_stopwords('my name is saksham'))
```

```python
import nltk
from nltk.corpus import stopwords

# 1. Download and convert to a SET once (outside the loop)
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# 2. Optimized function using list comprehension
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])

# 3. Apply it to the dataframe
df['review'] = df['review'].apply(remove_stopwords)
df.head()
```

```python
#THIS WILL DEFINATELY REMOVE THE EMOGI'S
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

remove_emoji("Loved the movie. It was 😁😡")
```

```python
# =====================================================================
# TEXT PREPROCESSING: TOKENIZATION & NLP LIBRARIES
# =====================================================================

# ---------------------------------------------------------------------
# 1. BASIC WORD TOKENIZATION
# Theory: Splitting a text string into individual words or tokens.
# The standard .split() function defaults to splitting by whitespace.
# ---------------------------------------------------------------------
sent1 = 'I am going to delhi'
words1 = sent1.split()
print("Word Tokenization Output:", words1)
# Expected Output: ['I', 'am', 'going', 'to', 'delhi']


# ---------------------------------------------------------------------
# 2. SENTENCE TOKENIZATION
# Theory: Segmenting a paragraph or block of text into individual sentences.
# Splitting purely on periods ('.') can cause issues if punctuation varies.
# ---------------------------------------------------------------------
sent2 = "I am going to delhi. I will stay there for 3 days. Let's hope the trip to be great"
sentences2 = sent2.split('.')
print("\nSentence Tokenization Output:", sentences2)
# Expected Output: ['I am going to delhi', ' I will stay there for 3 days', " Let's hope the trip to be great"]


# ---------------------------------------------------------------------
# 3. PROBLEMS WITH BASIC SPLIT FUNCTION
# Theory: The basic string .split() method has major limitations.
# - In sent3: The exclamation mark sticks to the word ('delhi!'), treating it as one token.
# - In sent4: It fails to split sentences cleanly because there is a '?' instead of a '.'.
# ---------------------------------------------------------------------
sent3 = 'I am going to delhi!'
print("\nProblem with split (Punctuation attached):", sent3.split())
# Expected Output: ['I', 'am', 'going', 'to', 'delhi!']

sent4 = 'Where do think I should go? I have 3 day holiday'
print("Problem with split (Missed question mark):", sent4.split('.'))
# Expected Output: ['Where do think I should go? I have 3 day holiday']


# ---------------------------------------------------------------------
# 4. ADVANCED TOKENIZATION WITH SPACY
# Theory: spaCy uses sophisticated language models (like 'en_core_web_sm')
# to handle complex tokens flawlessly. It correctly identifies text features
# like abbreviations (Ph.D), email addresses, distances, and currency symbols
# without breaking them apart incorrectly or leaving punctuation attached.
# ---------------------------------------------------------------------
import spacy

# Load the small English language model pipeline
nlp = spacy.load('en_core_web_sm')

# Defining complex test cases
sent5 = 'I have a Ph.D in A.I'
sent6 = "We're here to help! mail us at nks@gmail.com"
sent7 = 'A 5km ride cost $10.50'

# Processing strings through the spaCy NLP pipeline
doc1 = nlp(sent5)
doc2 = nlp(sent6)
doc3 = nlp(sent7)
doc4 = nlp(sent1)

print("\n--- Iterating through spaCy Tokens (doc4) ---")
for token in doc4:
    print(token.text)
```
