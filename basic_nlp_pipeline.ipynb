```python
import numpy as np

class MinimalNLPPipeline:
    def __init__(self):
        self.vocabulary = {}
        self.inverse_vocabulary = {}

    def fit_vocabulary(self, corpus):
        """Creates a unique deterministic mapping of words to numerical indices."""
        unique_words = sorted(list(set([word.lower().strip(".,!?") for sentence in corpus for word in sentence.split()])))

        for idx, word in enumerate(unique_words):
            self.vocabulary[word] = idx
            self.inverse_vocabulary[idx] = word
        return self

    def tokenize(self, text):
        """ yaha pe ham sentence ko slpit karke tokens ke format me banate hai """
        """Splits a raw string into standard discrete string tokens."""
        return [word.lower().strip(".,!?") for word in text.split()]
        # return  [word.lower().strip(".,!?") fo word in text.split()]

    def vectorize_one_hot(self, text):
        """Maps tokens to their exact spatial index in a one-hot matrix representation."""
        tokens = self.tokenize(text)
        num_tokens = len(tokens)
        vocab_size = len(self.vocabulary)

        # Initialize a matrix of shape (Sequence Length, Vocabulary Size)
        one_hot_matrix = np.zeros((num_tokens, vocab_size))

        for i, token in enumerate(tokens):
            if token in self.vocabulary:
                idx = self.vocabulary[token]
                one_hot_matrix[i, idx] = 1.0

        return one_hot_matrix

# --- Execution ---
corpus = [
    "Deep learning models scale beautifully.",
    "Transformers are modern deep learning architectures."
]

pipeline = MinimalNLPPipeline()
pipeline.fit_vocabulary(corpus)

sample_text = "Deep learning architectures scale."
vectorized_matrix = pipeline.vectorize_one_hot(sample_text)

print("Vocabulary Size:", len(pipeline.vocabulary))
print("One-Hot Tensor Shape (Tokens x Vocab Size):", vectorized_matrix.shape)
```
